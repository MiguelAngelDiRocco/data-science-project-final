## **Filtrado de Outliers**

**Justificación del tratamiento de outliers**

En este análisis se consideraron dos enfoques distintos para la detección de valores atípicos: el uso de **umbrales fijos** y el método estadístico del **rango intercuartílico (IQR)**. Ambos se representan mediante boxplots para comparar cómo cada criterio identifica los outliers.  

**Boxplots con umbrales fijos**
Se establecieron umbrales manuales basados en criterios físicos y realistas:  
- **Velocidad del viento**: 150 kph, correspondiente a un huracán de categoría 1 en la escala Saffir-Simpson.  
- **PM2.5**: 500 µg/m³, considerado un nivel extremadamente peligroso para la salud según la OMS.  

Este enfoque no depende de la dispersión estadística, sino que se fundamenta en **condiciones límite del mundo real**, de modo que se eliminan únicamente valores que son muy poco probables o físicamente irreales.  

**Boxplots con IQR**

Su principal ventaja es que se adapta a la distribución de los datos sin necesidad de supuestos externos. Sin embargo, en datasets globales como este, tiende a ser demasiado restrictivo, etiquetando como outliers valores que sí pueden ocurrir en contextos específicos (ej. tormentas intensas o ciudades con alta contaminación).  

**Elección del criterio**  
Aunque se muestran ambos enfoques para fines comparativos, se optó por trabajar con los **umbrales fijos**. Este criterio permite:  
- Mantener la coherencia con fenómenos físicos y límites reconocidos por organismos internacionales.  
- Evitar la eliminación de datos válidos que el IQR podría descartar de forma excesiva.  
- Asegurar que los modelos se entrenen con información representativa de situaciones reales, sin sesgarse por una definición puramente estadística de outlier.  

**En conclusión**, los boxplots con IQR se incluyen como referencia estadística, pero el análisis final se basa en los umbrales fijos, ya que reflejan mejor la realidad física y garantizan resultados más interpretables en un contexto global.  


```python
# IQR

# Velocidad del viento IQR
Q1_wind = df['wind_kph'].quantile(0.25)
Q3_wind = df['wind_kph'].quantile(0.75)
IQR_wind = Q3_wind - Q1_wind

lim_inf_wind = Q1_wind - 1.5 * IQR_wind
lim_sup_wind = Q3_wind + 1.5 * IQR_wind

print("Velocidad del viento - IQR:", IQR_wind)
print("Límite inferior:", lim_inf_wind)
print("Límite superior:", lim_sup_wind)

outliers_wind = df[(df['wind_kph'] < lim_inf_wind) | (df['wind_kph'] > lim_sup_wind)]
print("Cantidad de outliers (viento):", len(outliers_wind))

# PM2.5 IQR
Q1_pm25 = df['air_quality_PM2.5'].quantile(0.25)
Q3_pm25 = df['air_quality_PM2.5'].quantile(0.75)
IQR_pm25 = Q3_pm25 - Q1_pm25

lim_inf_pm25 = Q1_pm25 - 1.5 * IQR_pm25
lim_sup_pm25 = Q3_pm25 + 1.5 * IQR_pm25

print("PM2.5 - IQR:", IQR_pm25)
print("Límite inferior:", lim_inf_pm25)
print("Límite superior:", lim_sup_pm25)

outliers_pm25 = df[(df['air_quality_PM2.5'] < lim_inf_pm25) | (df['air_quality_PM2.5'] > lim_sup_pm25)]
print("Cantidad de outliers (PM2.5):", len(outliers_pm25))

# Creo el dataset df_iqr para hacer la comparacion con umbral fijo
df_iqr = df[
    (df['wind_kph'] >= lim_inf_wind) & (df['wind_kph'] <= lim_sup_wind) &
    (df['air_quality_PM2.5'] >= lim_inf_pm25) & (df['air_quality_PM2.5'] <= lim_sup_pm25)]

print("Tamaño original del dataset:", len(df))
print("Tamaño después de quitar outliers con IQR:", len(df_iqr))
```

**Comparación entre boxplots con umbral fijo e IQR**

En este análisis se aplicaron dos enfoques diferentes para la detección y tratamiento de outliers:

- **Umbral fijo:**  
  Se establecieron valores de referencia basados en criterios prácticos y de sentido común:
  - Velocidad del viento: 150 kph  
  - PM2.5: 500 µg/m³  
  Estos umbrales buscan reflejar límites físicos o realistas que difícilmente deberían superarse en condiciones normales.  

- **Rangos intercuartílicos (IQR):**  
  Método estadístico que calcula límites a partir de la distribución de los datos:  
  - Límite inferior = Q1 - 1.5 × IQR  
  - Límite superior = Q3 + 1.5 × IQR  
  Este enfoque detecta valores atípicos de manera automática según la dispersión de los datos.  

La comparación visual mediante boxplots permite observar cómo cambia la detección de outliers dependiendo del criterio aplicado. Mientras que el método IQR se adapta estrictamente a la estadística del dataset, los umbrales fijos aportan un control basado en criterios físicos y de contexto.

# **Algoritmo de regresion**

Se eligió un modelo de regresión porque la variable objetivo de este proyecto (log_air_quality_PM2.5) es un valor numérico continuo.
El objetivo del análisis es poder predecir la concentración de contaminantes en el aire a partir de diferentes variables (como temperatura, humedad, viento, país, etc.).

En este caso, no se busca clasificar en categorías (por ejemplo, "bajo", "medio" o "alto"), sino estimar directamente un número real, lo que corresponde a un problema de regresión.
Por eso se entrenaron algoritmos de regresión (Regresión Lineal y Random Forest Regressor) que permiten ajustar un modelo y luego evaluar la precisión de las predicciones mediante métricas como MAE, RMSE y R².

## **Importando librerias para el modelo**

```python
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import r2_score
```

## **Seleccionando las variables**

```python
# Variables predictoras (X) y variable objetivo (y)
X = df_ml.drop(columns=['log_air_quality_PM2.5'])
y = df_ml['log_air_quality_PM2.5']
```

## **Separando dataset en train y test**

```python
# Separar en train y test
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=2)

print(X_train.shape, X_test.shape)
```

# **Entrenamiento modelo Linear Regression**

```python
# crear el modelo
reg_lineal = LinearRegression()
# Ajustar el modelo con X_train y y_train
reg_lineal.fit(X_train, y_train)
# Predecir con X_test
y_pred = reg_lineal.predict(X_test)
```

## **Metricas de Linear Regression**

```python
print("Resultados de la regresion lineal:")
# MAE
print("MAE:", mean_absolute_error(y_test, y_pred))
# MSE
print("MSE:", mean_squared_error(y_test, y_pred))
# RMSE
print("RMSE:", np.sqrt(mean_squared_error(y_test, y_pred)))
# R²
print("R²:", r2_score(y_test, y_pred))
```

## **Conclusión Linear Regression**

El modelo de Regresión Lineal logró predecir la variable objetivo con un error medio (MAE) de aproximadamente 0.61, un MSE de 0.60 y un RMSE cercano a 0.77. Estos valores muestran que las predicciones no son perfectas, pero sí se acercan en promedio a los valores reales.

El coeficiente de determinación (R² ≈ 0.45) indica que el modelo explica alrededor del 45% de la variabilidad de los datos. Esto significa que existe una capacidad predictiva aceptable, aunque también queda margen de mejora con algoritmos más complejos como Random Forest u otros métodos no lineales.

En conclusión, la regresión lineal funciona como un primer modelo base para este problema, pero es probable que un modelo más flexible capture mejor las relaciones entre las variables.

# **Entrenamiento modelo Random Forest Regressor**

```python
# crear el modelo
rf_reg = RandomForestRegressor(n_estimators=100, random_state=2)
# Ajustar el modelo con X_train y y_train
rf_reg.fit(X_train, y_train)
# Predecir con X_test
y_pred_rf = rf_reg.predict(X_test)
```

## **Metricas de Random Forest Regressor**

```python
print("Resultados del Random Forest Regressor:")
# MAE
print("MAE:", mean_absolute_error(y_test, y_pred_rf))
# MSE
print("MSE:", mean_squared_error(y_test, y_pred_rf))
# RMSE
print("RMSE:", np.sqrt(mean_squared_error(y_test, y_pred_rf)))
# R²
print("R²:", r2_score(y_test, y_pred_rf))
```

## **Conclusión Random Forest Regressor**

El modelo de Random Forest Regressor mostró un mejor desempeño en comparación con la regresión lineal. El MAE (0.45), el MSE 0.35 y el RMSE (0.62) son menores que en el modelo lineal, lo que indica que las predicciones del Random Forest están más cerca de los valores reales.

Además, el R² = 0.67 significa que este modelo logra explicar aproximadamente el 67% de la variabilidad de los datos, lo cual representa una mejora significativa respecto al 45% que explicaba la regresión lineal.

En conclusión, el Random Forest resulta un modelo más preciso y robusto para este problema, mostrando una mejor capacidad de generalización y mayor ajuste a los datos.
